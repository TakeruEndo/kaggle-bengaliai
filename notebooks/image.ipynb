```python
train = pd.read_csv('../input/train.csv')
data0 = pd.read_feather('../input/train_image_data_0.feather')
```

```python
data_a = data0.iloc[0][1:].values.reshape(137,236).astype(np.float)
```

```python
data_a.shape
```

```python
image = cv2.resize(data_a,(64,64))
```

```python
im_list = np.asarray(image)
plt.imshow(im_list)
plt.show()
```

```python
import albumentations as albu
# train用のデータ拡張
data_transforms = albu.Compose([
    albu.HorizontalFlip(p=0.5),
    albu.VerticalFlip(p=0.5),
    albu.Resize(32, 32),
])
```

```python
augmented = data_transforms(image=image)
```

```python
image = augmented['image']
```

```python
im_list = np.asarray(image)
plt.imshow(im_list)
plt.show()
```
